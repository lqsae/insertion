```python
Rhg1='G://工作目录/河南省农业科学院1个大豆样品三代变异检测分析技术服务（委托）合同/dadou.Rhg1.mapinfo'
Rhg2='G://工作目录/河南省农业科学院1个大豆样品三代变异检测分析技术服务（委托）合同/dadou.Rhg4.mapinfo'
```

```python
insertion='Rhg1'
all_break_point=[]
insertion_list=[]
ref_list=[]
query_name_list=[]


with open(Rhg1) as f1:
    for line in f1:
        tags=line.strip().split('\t')
        list1=tags[:-1]
        query=tags[-1]
        list1.sort(key=lambda x: int(x.split('_')[-1]))
        length=len(list1)
        
        
        for i,j in enumerate(list1):
            chr=j.split('_')[0]
            ins_start=int(j.split('_')[-2])
            ins_end=int(j.split('_')[-1])
            
            
            if chr==insertion and i==0:
                right_start=int(list1[i+1].split('_')[-2])
                right_end=int(list1[i+1].split('_')[-1])
                if right_start-50<=ins_end<=right_start+50:
                    break_point=int(list1[i+1].split('_')[-4])

                    ref_list.append(list1[i+1])
                    insertion_list.append(j)
                    query_name_list.append(query)
                    all_break_point.append(break_point)
                    
                    print(list1[i+1],j,break_point,query)
              
            
            elif chr==insertion and i==length-1:
                left_start=int(list1[i-1].split('_')[-2])
                left_end=int(list1[i-1].split('_')[-1])
                if left_end-50<=ins_start<=left_end+50:
                    break_point=int(list1[i-1].split('_')[-3])
                    
                    ref_list.append(list1[i-1])
                    insertion_list.append(j)
                    query_name_list.append(query)
                    all_break_point.append(break_point)
                    
                    print(list1[i-1],j,break_point,query)
                    
            elif chr==insertion and (i not in [0,length-1]):
                right_start=int(list1[i+1].split('_')[-2])
                right_end=int(list1[i+1].split('_')[-1])
                left_start=int(list1[i-1].split('_')[-2])
                left_end=int(list1[i-1].split('_')[-1])
                if right_start-50<=ins_end<=right_start+50:
                    break_point=int(list1[i+1].split('_')[-4])
                    ref_list.append(list1[i+1])
                    insertion_list.append(j)
                    query_name_list.append(query)
                    all_break_point.append(break_point)
                    print(list1[i+1],j,break_point,query)
                    
                elif left_end-50<=ins_start<=left_end+50:
                    break_point=int(list1[i-1].split('_')[-3])
                    ref_list.append(list1[i-1])
                    insertion_list.append(j)
                    query_name_list.append(query)
                    all_break_point.append(break_point)
                    
                    print(list1[i-1],j,break_point,query)
        
```

```python
dict_info={'ref':ref_list,'insertion':insertion_list,'break_point':all_break_point,'query_name':query_name_list}
```
